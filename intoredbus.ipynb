```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
dfandhra=pd.read_csv("andhra_bus.csv")
dfandhra
```

```python

driver_andhra_bus = webdriver.Chrome()
Bus_names_andhra_bus = []
Bus_types_andhra_bus = []
Departure_andhra_bus = []
Arrival_andhra_bus = []
Ratings_andhra_bus = []
Total_Duration_andhra_bus = []
Prices_andhra_bus = []
Seats_Available_andhra_bus = []
Route_names = []
Route_links = []


for i,r in dfandhra.iterrows(): 
    link=r["Route_link"]  
    routes=r["Route_name"] 

# Loop through each link
    driver_andhra_bus.get(link)
    time.sleep(2)   

    # Click on elements to reveal bus details
    elements = driver_andhra_bus.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]") 
    for element in elements: 
        element.click()  
        time.sleep(2)  
        
    # click elements to views bus
    try:
        clicks = driver_andhra_bus.find_element(By.XPATH, "//div[@class='button']")
        clicks.click() 
    except:
        continue  
    time.sleep(2)
    
    scrolling = True
    while scrolling:
        old_page_source = driver_andhra_bus.page_source   
        
        # Use ActionChains to perform a PAGE_DOWN
        ActionChains(driver_andhra_bus).send_keys(Keys.PAGE_DOWN).perform() 
                                                                    
        
        time.sleep(5)  
        
        new_page_source = driver_andhra_bus.page_source  
        
        if new_page_source == old_page_source:
            scrolling = False                   

    # Extract bus details
    bus_name = driver_andhra_bus.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")   
    bus_type = driver_andhra_bus.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")  
    start_time = driver_andhra_bus.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_time = driver_andhra_bus.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    total_duration = driver_andhra_bus.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    
    price = driver_andhra_bus.find_elements(By.XPATH, '//div[@class="fare d-block"]//span')
    seats = driver_andhra_bus.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    try:
        rating = driver_k.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except:
        continue
    # Append data to respective lists
    for bus in bus_name:
        Bus_names_andhra_bus.append(bus.text)
        Route_links.append(link)
        Route_names.append(routes)
    for bus_type_elem in bus_type:
        Bus_types_andhra_bus.append(bus_type_elem.text)
    for start_time_elem in start_time:
        Departure_andhra_bus.append(start_time_elem.text)
    for end_time_elem in end_time:
        Arrival_andhra_bus.append(end_time_elem.text)
    for total_duration_elem in total_duration:
        Total_Duration_andhra_bus.append(total_duration_elem.text)
    for ratings in rating:
        Ratings_andhra_bus.append(ratings.text)
    for price_elem in price:
        Prices_andhra_bus.append(price_elem.text)
    for seats_elem in seats:
        Seats_Available_andhra_bus.append(seats_elem.text)
        
print("Successfully Completed")
```

```python
# from list to convert data frame
data = {
    'Bus_name': Bus_names_andhra_bus,
    'Bus_type': Bus_types_andhra_bus,
    'Start_time': Departure_andhra_bus,
    'End_time': Arrival_andhra_bus,
    'Total_duration': Total_Duration_andhra_bus,
    'Price': Prices_andhra_bus,
    "Seats_Available":Seats_Available_andhra_bus,
    "Ratings":Ratings_andhra_bus,
    'Route_name': Route_names,
    'Route_link': Route_links
}

df_bus_andhra = pd.DataFrame(data)
path= r"C:\Redbus\df_bus_andhra.csv"
df_bus_andhra.to_csv(path,index=False)
```
